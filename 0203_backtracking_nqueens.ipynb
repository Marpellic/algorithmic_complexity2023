```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showBoard(board):
    n = len(board)
    b = np.zeros((n, n, 3), dtype=int)
    b[:, :] = [200, 100, 50]
    b[::2, ::2] = [250, 150, 50]
    b[1::2, 1::2] = [250, 150, 50]
    plt.imshow(b)
    for row, col in enumerate(board):
        if col == -1: continue
        plt.text(col, row, u"\u265B", size=200/n, ha='center', va='center')
    plt.axis("off")
    plt.show()

showBoard([1, 3, 0, 2, 5, 7, 4, 6])
```

```python
 def validPosition(board, row, col):
    for rowi in range(row):
        delta = row - rowi
        if board[rowi] == col or \
            board[rowi] == col - delta or \
            board[rowi] == col + delta:
            return False
    return True
```

```python
def nqueens(board, row):
    n = len(board)
    if row == n:
        showBoard(board)
        return

    for col in range(n):
        if validPosition(board, row, col):
            board[row] = col
            nqueens(board, row + 1)
```

```python
n = 4
nqueens([-1]*n, 0)
```
