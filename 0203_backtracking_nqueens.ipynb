# [N-Queens](https://en.wikipedia.org/wiki/Eight_queens_puzzle)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showBoard(board: list):
    n = len(board)
    b = np.zeros((n, n, 3), dtype=int)
    b[:, :] = [230, 150, 130]
    b[::2, ::2] = [250, 210, 180]
    b[1::2, 1::2] = [250, 210, 180]
    plt.imshow(b)
    for row, col in enumerate(board):
        if col != -1:
            plt.text(col, row, u"\u265B", size=200/n, ha='center', va='center')
    plt.axis("off")
    plt.show()

showBoard([1, 3, 0, 2, 5, 7, 4, 6])
```

```python
def validPosition(board, row, col):
    for ri in range(row):
        d = row - ri
        if board[ri] == col or board[ri] == col - d or board[ri] == col + d:
            return False
    return True
```

```python
def nqueens(board, row):
    n = len(board)
    if row == n:
        showBoard(board)
        return

    for col in range(n):
        if validPosition(board, row, col):
            board[row] = col
            nqueens(board, row + 1)
```

```python
n = 4
nqueens([-1]*n, 0)
```

```python
def nqueens(board, row):
    n = len(board)
    if row == n:
        showBoard(board)
        return True

    for col in range(n):
        if validPosition(board, row, col):
            board[row] = col
            if nqueens(board, row + 1):
                return True
    return False
```
