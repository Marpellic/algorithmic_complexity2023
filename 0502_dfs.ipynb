```python
with open('1.g') as f:
    G = []
    for line in f:
        G.append([int(v) for v in line.split()])

print(G)
```

```python
import graphviz as gv

dot = gv.Graph("aca")
checklist = set()
for u in range(len(G)):
    for v in G[u]:
        if not f"({u},{v})" in checklist and not f"({v},{u})" in checklist:
            dot.edge(str(u), str(v))
            checklist.add(f"({u},{v})")
            checklist.add(f"({v},{u})")

dot
```

```python
def dfs(G, s):
    n = len(G)
    visited = [False]*n
    path = [-1]*n

    def _dfs(u):
        visited[u] = True
        for v in G[u]:
            if not visited[v]:
                path[v] = u
                _dfs(v)

    _dfs(s)
    return path
```

```python
dfs(G, 0)
```
